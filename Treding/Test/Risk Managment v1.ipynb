# With Charge

```python
import random
import pandas as pd

# Initialize capital and other variables
capital = 20000
trades_per_day = 3
charge_per_trade = 50
max_loss = 0.3
# min_profit = 0.3
# min_profit = lambda: random.uniform(0.3, 1)
min_profit = lambda: 0.5

day_list = []
trade_list = []
tred_margin_list = []
status_list = []
profit_loss_list = []
profit_loss_percentage_list = []
tred_new_margin_list = []
capital_list = []
used_capital_list = []  # List to store Used_Capital

while capital >=  0 and capital <= 100000:  # Run until capital doubles or reaches zero
    capital_per_trade = capital / trades_per_day  # Divide capital equally among trades for the day

    for trade in range(1, trades_per_day + 1):
        # Simulate profit/loss using random choice
        is_profit = random.choice([True, False])  # True represents a profit, False represents a loss\
        is_profit = random.choices([True, False], weights=[0.5, 0.5])[0]

        if is_profit:
            profit = ((capital_per_trade) * min_profit()) - charge_per_trade
            status = 'Profit'
        else:
            profit = -capital_per_trade * max_loss - charge_per_trade
            status = 'Loss'

        tred_margin = capital_per_trade
        tred_new_margin = tred_margin + profit - charge_per_trade  # Deduct the charge from the new margin
        capital += profit 
        used_capital = capital - profit 

        # Append trade data to the lists
        day_list.append(f'Day {len(day_list) // trades_per_day + 1}')
        trade_list.append(f'Tred {trade}')
        tred_margin_list.append(tred_margin)
        status_list.append(status)
        profit_loss_list.append(profit)
        profit_loss_percentage_list.append((profit / tred_margin) * 100)
        tred_new_margin_list.append(tred_new_margin)
        capital_list.append(capital)
        used_capital_list.append(used_capital)

# Create a pandas dataframe with the desired column order
data = {
    'Day': day_list,
    'Trade': trade_list,
    'Used_Capital': used_capital_list,
    'Tred_Margin': tred_margin_list,
    'Status': status_list,
    'Profit Loss': profit_loss_list,
    'Profit Loss%': profit_loss_percentage_list,
    'Tred_New_Margin': tred_new_margin_list,
    'Capital': capital_list,
}
df = pd.DataFrame(data)


if capital <= 0:
    print("Loss")
else:
    print("Target Achieved.")
    
print("Total Day ",df.shape[0]//3)
print("Total Tred ",df.shape[0])
loss_count = df[df['Status'] == 'Loss'].shape[0]
profit_count = df[df['Status'] == 'Profit'].shape[0]
print(f"Loss Count: {loss_count}")
print(f"Profit Count: {profit_count}")
# df.to_csv("Risk_management.csv")
df
```
