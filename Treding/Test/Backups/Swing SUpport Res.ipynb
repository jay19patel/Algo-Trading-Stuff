```python
import pandas as pd
from ta import momentum, trend
from ta import momentum,trend
import numpy as np
import yfinance as yf
```

```python
# Define the stock symbol and timeframe
symbol = '^NSEI'  # Replace 'AAPL' with the symbol of the stock you want to fetch.
# symbol = 'HDFCBANK.NS'  # Replace 'AAPL' with the symbol of the stock you want to fetch.
interval = '1d'
period = '50d'

df = yf.download(symbol, interval=interval, period=period)
df['Date'] = df.index.date
df['Time'] = df.index.time
df = df.reset_index(drop=True)
df = df.round(2)
df
```

# EMA AND SMA AND CANDLE

```python
support_resistance = []
df['10EMA'] = df['Close'].ewm(span=10, adjust=False).mean()
current_high = df.tail(100)['High'].max()
current_low = df.tail(100)['Low'].min()
trend = None
ema_values = df['10EMA']

change_error = 10 # Number of consecutive values to consider for trend change

for i in range(1, len(ema_values) - 1):
    if ema_values[i] > ema_values[i - 1] and ema_values[i] > ema_values[i + 1]:
        if i + change_error < len(ema_values) and all(ema_values[i] > ema_values[j] for j in range(i + 1, i + change_error)):
            trend = "Resistance"
    elif ema_values[i] < ema_values[i - 1] and ema_values[i] < ema_values[i + 1]:
        if i + change_error < len(ema_values) and all(ema_values[i] < ema_values[j] for j in range(i + 1, i + change_error)):
            trend = "Support"
    else:
        trend = ""

    if trend:
        support_resistance.append([ema_values[i], df['Date'].iloc[i], df['Time'].iloc[i], trend])

support_resistance_df = pd.DataFrame(support_resistance, columns=['10EMA', 'Date', 'Time', 'Support/Resistance'])

# support_resistance_df.to_csv("nifty_50_support.resis.csv")
support_resistance_df

```

```python
import yfinance as yf
import plotly.graph_objects as go

# Define the stock symbol and timeframe
symbol = '^NSEI'
interval = '5m'
period = '1d'

# Download stock data
df = yf.download(symbol, interval=interval, period=period)
df['Date'] = df.index.date
df['Time'] = df.index.time
df = df.reset_index(drop=True)
df = df.round(2)

# Calculate maximum and minimum prices for the last 50 candles
max_price = df['High'].tail(50).max()
min_price = df['Low'].tail(50).min()

# Create Fibonacci retracement levels
fibonacci_levels = [0, 0.706, 0.618,0.79,1]

# Calculate Fibonacci retracement values
fibonacci_values = [min_price + level * (max_price - min_price) for level in fibonacci_levels]
```
