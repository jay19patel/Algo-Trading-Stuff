```python
import random
import pandas as pd



def simulate_trading():
    Capital=5000
    Tred_per_day=3
    Max_loss=0.2
    Min_Profit=0.3
    Charges=50

    
    all_day_dataframes = []  # List to store DataFrames for each day
    Day = 1

    while 0 <= Capital <= 100000:
        # print("CAPITAL :", Capital, "Day :", Day)
        Day_margin = Capital * 0.7
        Tred_margin = Day_margin / 2
        day_tred_logs = []

        # Generate trade logs for the day
        day_data = []  # List to store data for each trade in the day
        for i in range(1, Tred_per_day + 1):
            is_profit = random.choices([True, False], weights=[0.6, 0.4])[0]
            if is_profit:
                status = "Profit"
                pnl = Tred_margin * Min_Profit - Charges
            else:
                status = "Loss"
                pnl = -Tred_margin * Max_loss - Charges
            day_tred_logs.append(pnl)

            # Append data for each trade to the day_data list
            day_data.append({
                "Day": f"Day {Day}",
                "Tred": f"Tred {i}",
                "Capital": Capital,
                "Day Margin": Day_margin,
                "Tred Margin": Tred_margin,
                "Status": status,
                "P&L": pnl,
                "P&L(%)": round((pnl / Tred_margin) * 100, 2),
                "Day PNL": sum(day_tred_logs),
                "New Capital": Capital + sum(day_tred_logs),
            })

        # Create a DataFrame for the day's trade logs
        day_dataframe = pd.DataFrame(day_data)

        # Append the day's DataFrame to the list
        all_day_dataframes.append(day_dataframe)

        Day += 1
        Capital = round(Capital + sum(day_tred_logs) - Charges, 2)  # Deduct charges

    # Concatenate all DataFrames in the list to create the final trade logs DataFrame
    Tred_logs = pd.concat(all_day_dataframes, ignore_index=True)
    return Tred_logs["New Capital"].iloc[-1]  # Return the final capital

# # Example usage
# final_capital = simulate_trading()
# print("Final Capital:", final_capital)
# Tred_logs
```

```python
count_list = []
for i in range(100):
    st = simulate_trading()
    if st <= 500:
        count_list.append("Loss")
    else:
        count_list.append("Target")
loss_count, target_count = count_list.count('Loss'), count_list.count('Target')
print(f'Loss Count: {loss_count}, Target Count: {target_count}')
print("Accurcy :",100*target_count/100)
```
