```python

def get_stock_info(symbol):
    stock = yf.Ticker(symbol)
    return stock.info
```

```python
df = pd.DataFrame([])
```

```python
import yfinance as yf
from datetime import datetime, timedelta

def calculate_outperformance(sbi_symbol, start_date=None, end_date=None):
    start_date = start_date or (datetime.now() - timedelta(days=31)).strftime('%Y-%m-%d')
    end_date = end_date or datetime.now().strftime('%Y-%m-%d')
    nifty_symbol = '^NSEI'

    sbi_data = yf.download(sbi_symbol, start=start_date, end=end_date)['Adj Close'].pct_change().add(1).cumprod()
    nifty_data = yf.download(nifty_symbol, start=start_date, end=end_date)['Adj Close'].pct_change().add(1).cumprod()

    outperformance = (sbi_data.iloc[-1] / nifty_data.iloc[-1] - 1) * 100
    return outperformance

```

```python
for i in symbol_list:
  symbol = f"{i}.BO"
  try:
      data = get_stock_info(symbol)
      market_cap = data["marketCap"]
      name = data['shortName']
      industry = data['industry']
      sector = data['sector']

      price = data['currentPrice']
      daily_volume = data['volume']
      average_volume_10days = data['averageVolume10days']
      week_change_52 = data['52WeekChange']
      fifty_two_week_high = data['fiftyTwoWeekHigh']
      away_from_52_days_high = abs(((price - fifty_two_week_high) / fifty_two_week_high) * 100 if fifty_two_week_high != 0 else float('inf'))
      day_average_50 = data['fiftyDayAverage']
      day_average_200 = data['twoHundredDayAverage']
      float_shares = data['floatShares']
      shares_outstanding = data['sharesOutstanding']
      change_52_weeks = data['52WeekChange']
      recommendationKey = data['recommendationKey']
      outperformance = calculate_outperformance(symbol)

      # Create a DataFrame for each symbol
      n_df = pd.DataFrame({
          'Symbol': [i],
          'Name': [name],
          'Industry': [industry],
          'Sector': [sector],
          'MarketCap': [market_cap],
          'Price': [price],
          'DailyVolume': [daily_volume],
          'AverageVolume10Days': [average_volume_10days],
          '52WeekChange': [week_change_52],
          'FiftyTwoWeekHigh': [fifty_two_week_high],
          'AwayFrom52WeekHigh': [away_from_52_days_high],
          'FiftyDayAverage': [day_average_50],
          'TwoHundredDayAverage': [day_average_200],
          'FloatShares': [float_shares],
          'SharesOutstanding': [shares_outstanding],
          'Change52Weeks': [change_52_weeks],
          'Predicted_status':[recommendationKey],
          'Outperformance': [outperformance]
      })
      df = pd.concat([df, n_df], axis=0, ignore_index=True)
  except Exception as e:
        print(f"Error fetching data for {symbol}: {e}")

```
